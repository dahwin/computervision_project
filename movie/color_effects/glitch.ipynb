```python
import cv2
import numpy as np
import random

# Load your video
cap = cv2.VideoCapture('fancy.mp4')

# Get the frame width and height
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Create a window to display the video with the same size as the video frames
cv2.namedWindow("Glitch Video", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Glitch Video", frame_width, frame_height)

# Loop through the video frames
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Apply a random glitch effect to each frame
    if random.random() < 0.2:  # Increase the glitch probability
        # Generate random coordinates for glitch effect
        x1, x2 = sorted(np.random.randint(0, frame.shape[1], size=2))
        y1, y2 = sorted(np.random.randint(0, frame.shape[0], size=2))

        # Ensure positive dimensions
        x2 = max(x1 + 1, x2)
        y2 = max(y1 + 1, y2)

        # Apply an intense glitch effect with pixel noise
        frame[y1:y2, x1:x2] = np.random.randint(0, 256, size=(y2 - y1, x2 - x1, 3), dtype=np.uint8)

        # Apply distortion by swapping random image regions
        x1, x2 = sorted(np.random.randint(0, frame.shape[1], size=2))
        y1, y2 = sorted(np.random.randint(0, frame.shape[0], size=2))
        frame[y1:y2, x1:x2] = cv2.resize(frame[y1:y2, x1:x2], (x2 - x1, y2 - y1))

    # Display the modified frame
    cv2.imshow("Glitch Video", frame)

    # Press 'q' to exit the loop
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

# Release video object and close windows
cap.release()
cv2.destroyAllWindows()
```
