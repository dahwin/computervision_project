```python
import cv2
import matplotlib.pyplot as plt

def draw_yolo_boxes(img, annotation_path, class_names):
    with open(annotation_path, 'r') as f:
        annotations = f.readlines()
        print(annotations)

    for ann in annotations:
        cls, x, y, w, h = map(float, ann.split())
        cls_id = int(cls)
        print(x,y,w,h)
        # Denormalize coordinates
        width, height = img.shape[1], img.shape[0]
        print(width,height)
        x1 = int((x - w / 2) * width)
        y1 = int((y - h / 2) * height)
        x2 = int((x + w / 2) * width)
        y2 = int((y + h / 2) * height)
        print(x1,y1,x2,y2)
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Blue rectangle
        print(cls_id)
        cv2.putText(img, class_names[cls_id], (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

image_path =  r"C:\Users\ALL USER\Desktop\e\all_data\blackhole\test\44.png"

# annotation_path = image_path.replace('.png','.txt')
annotation_path = r"C:\Users\ALL USER\Desktop\e\all_data\blackhole\test\44.txt"



# image_path = r"C:\Users\ALL USER\Downloads\4930.png"
# annotation_path = r"C:\Users\ALL USER\Desktop\e\all_data\datasets_new\bbox_converted.txt"
class_names = ['circel minimize','circel maximize and down','circel close','specker','minus']
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Matplotlib works in RGB

draw_yolo_boxes(img, annotation_path, class_names)

plt.imshow(img)
plt.show()
```
