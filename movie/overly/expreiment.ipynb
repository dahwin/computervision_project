```python
import moviepy.editor as mp

img = '1.jpeg'

# Load the image
slide = mp.ImageClip(img)
# Divide each element in the list by 2
resolution = [x // 2 for x in slide.size]
# Set the duration and frame rate
slide = slide.set_duration(10).set_fps(24)
slide = slide.resize(height=resolution[1],width=resolution[0])
# Define a custom resizing function for the zoom-out effect
def resize_func(t):
    # Scale down the image only in the last 2 seconds of the video
    if t <= 8:  # Scale factor is 1 from the start to 8 seconds
        zoom_factor = 1.0
    else:
        # Scale down the image gradually in the last 2 seconds
        zoom_factor = 1 - 0.5 * (t - 8)  # Adjust the factor as needed
        
    return zoom_factor

# Apply the custom resizing function
slide = slide.resize(resize_func)

# Center the image
slide = slide.set_position(('center', 'center'))

# Create the video clip
slide = mp.CompositeVideoClip([slide])

# Write the video file
slide.write_videofile('zoom-out.mp4')
```

```python
import moviepy.editor as mp

img = '1.jpeg'

# Load the image
slide = mp.ImageClip(img)

# Divide each element in the list by 2
resolution = [x // 2 for x in slide.size]
print(resolution)
# Set the duration and frame rate
slide = slide.set_duration(10).set_fps(24)


# Define a custom resizing function for the zoom-in effect
def resize_func(t):
    if t <= 0.5:
        # Zoom in from small to full size during the first half of the video
        zoom_factor = 0.1 + 0.9 * (t * 2)
    else:
        # Keep it at full size during the second half of the video
        zoom_factor = 1.0
    print(zoom_factor)
    return zoom_factor

# Apply the custom resizing function
slide2 = slide.resize(resize_func)

# Center the image
slide3 = slide2.set_position(('center', 'center'))

# Create the video clip
slide1 = mp.CompositeVideoClip([slide3], size=resolution)

# Write the video file
slide1.write_videofile('zoom-test-1.mp4')
```
