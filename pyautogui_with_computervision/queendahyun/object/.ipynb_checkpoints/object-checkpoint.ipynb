```python
import requests
base =  "https://d28a-35-221-210-134.ngrok-free.app"
url = f"{base}/process-image/"
data = {
    "filter_range": "top_right_corner",
    "object": "restore down",
    "close": None,
    "direction": None,
    "n": 1,
    "verify": False
}

response = requests.post(url, json=data)
if response.status_code == 200:
    with open("processed_image.png", "wb") as f:
        f.write(response.content)
    print("Image saved as processed_image.png")
else:
    print(f"Error: {response.json()}")
```

```python
import requests
from PIL import Image
import io

base = "https://d28a-35-221-210-134.ngrok-free.app"
url = f"{base}/process-image/"
data = {
    "filter_range": "top_right_corner",
    "object": "scrollbar",
    "close": None,
    "direction": None,
    "n": 1,
    "verify": True
}

response = requests.post(url, json=data)
if response.status_code == 200:
    # Load the image into a PIL Image object
    pil_image = Image.open(io.BytesIO(response.content))
    
else:
    print(f"Error: {response.json()}")
pil_image
```
