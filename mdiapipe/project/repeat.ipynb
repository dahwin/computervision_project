```python
import cv2
import numpy as np

# Open the video files
cap1 = cv2.VideoCapture("C:\\Users\\Pc\\Desktop\\conputer_Vison\\mdiapipe\\project\\upwork\\Upwork_07_Chest_Open.mp4")
cap2 = cv2.VideoCapture("C:\\Users\\Pc\\Desktop\\conputer_Vison\\mdiapipe\\project\\upwork\\Upwork_07_Chest_Open_Anim Bubbles.mp4")

# Define the size of the rectangles
rectangle_size = (940, 1080)

# Create the window
cv2.namedWindow("Video", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Video", (2 * rectangle_size[0], rectangle_size[1]))

while(cap1.isOpened() and cap2.isOpened()):
    ret1, frame1 = cap1.read() 
    ret2, frame2 = cap2.read()
    
    if ret1 and ret2:
        # Resize the frames to fit the rectangles
        frame1 = cv2.resize(frame1, rectangle_size)
        frame2 = cv2.resize(frame2, rectangle_size)

        # Horizontally concatenate the frames
        merged_frame = np.hstack((frame1, frame2))

        # Display the merged frame
        cv2.imshow("Video", merged_frame)
    else:
        cap1.set(cv2.CAP_PROP_POS_FRAMES, 0)
        cap2.set(cv2.CAP_PROP_POS_FRAMES, 0)
        continue

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap1.release()
cap2.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
from PIL import Image
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```
