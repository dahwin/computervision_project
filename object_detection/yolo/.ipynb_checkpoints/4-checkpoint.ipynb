```python
from PIL import Image, ImageDraw, ImageFont
import os


# Function to add white number with black border
def add_number_with_border(image_path, n):
    # Open the original image
    original_image = Image.open(image_path)

    # Create a drawing object
    draw = ImageDraw.Draw(original_image)

    # Specify the font size
    font_size = 50

    # Load the TrueType font
    font = ImageFont.truetype("arial.ttf", font_size)

    # Define text position
    text_position = (10, 10)

    # Define text color
    border_color = "black"
    fill_color = "white"

    # Draw black border
    border_offset = 2
    for x_offset in range(-border_offset, border_offset + 1):
        for y_offset in range(-border_offset, border_offset + 1):
            draw.text((text_position[0] + x_offset, text_position[1] + y_offset), str(n), font=font, fill=border_color)

    # Draw white number
    draw.text(text_position, str(n), font=font, fill=fill_color)

    # Save the modified image to the same path
    original_image.save(image_path)

# Process each image
for n in range(0, 16):
    image_path = f'{folder_path}/last{n}.png'
    print(f"Processing {image_path}")

    add_number_with_border(image_path, n)

    print(f"Number {n} added to the image. Saved to {image_path}")
```

```python
# Efficient
import os
from PIL import Image

# Directory paths
directory_path = r'C:\Users\ALL USER\Desktop\e\all_data\blackhole\all_image'

# List all files in the directory
files = os.listdir(directory_path)

# Organize files into groups based on numbers
file_groups = {}
for file in files:
    base_name = os.path.basename(file)
    number = ''.join(filter(str.isdigit, base_name))
    if number not in file_groups:
        file_groups[number] = []
    file_groups[number].append(file)

# Process each group
for n, group_files in file_groups.items():
    image_path = f"{home_m}/image/last{n}.png"
    top = None
    bottom = None
    for file in group_files:
        file_path = os.path.join(directory_path, file)
        if "bottom" in file:
            bottom = file_path
        elif "top" in file:
            top = file_path

    # Open the split images
    top_half = Image.open(top)
    bottom_half = Image.open(bottom)

    # Get the dimensions of the split images
    width, height = top_half.size

    # Create a new image with twice the height of the split images
    reunified_image = Image.new("RGB", (width, height * 2))

    # Paste the split images onto the reunified image
    reunified_image.paste(top_half, (0, 0))
    reunified_image.paste(bottom_half, (0, height))

    # Save the reunified image
    reunified_image.save(image_path)
```

```python
# Load the original image to get its dimensions
original_image_path = 'full_hd.png'
original_image = Image.open(original_image_path)
original_width, original_height = original_image.size

# Reconstruct the top and bottom halves
top_half = Image.new('RGB', (original_width // 2, original_height // 2))
top_half.paste(Image.open(f"{home}/last1.png"), (0, 0))
top_half.paste(Image.open(f"{home}/last2.png"), (original_width // 4, 0))

bottom_half = Image.new('RGB', (original_width // 2, original_height // 2))
bottom_half.paste(Image.open(f"{home}/last3.png"), (0, 0))
bottom_half.paste(Image.open(f"{home}/last4.png"), (original_width // 4, 0))

# Reconstruct the left and right halves
left_half = Image.new('RGB', (original_width // 2, original_height))
left_half.paste(top_half, (0, 0))
left_half.paste(bottom_half, (0, original_height // 2))

right_half = Image.new('RGB', (original_width // 2, original_height))
right_half.paste(Image.open(f"{home}/last5.png"), (0, 0))
right_half.paste(Image.open(f"{home}/last6.png"), (original_width // 4, 0))
right_half.paste(Image.open(f"{home}/last7.png"), (0, original_height // 2))
right_half.paste(Image.open(f"{home}/last8.png"), (original_width // 4, original_height // 2))

# Merge the left and right halves to form the final image
reunified_image = Image.new('RGB', (original_width, original_height))
reunified_image.paste(right_half, (0, 0))
reunified_image.paste(left_half, (original_width // 2, 0))

# Save the reunified image
reunified_image.save(f"{home_m}/youtube_reunified.png")

print("Image reunified successfully!")
```

```python
import os

# Define the directory path
directory = r'C:\Users\ALL USER\Desktop\e\all_data\blackhole\all_image'

# Get the list of files in the directory
files = os.listdir(directory)

# Create absolute paths for each file
file_paths = [os.path.join(directory, file) for file in files]

# Sort the file paths based on their creation time
sorted_file_paths = sorted(file_paths, key=os.path.getctime)

# Print the absolute paths of the sorted files
for file_path in sorted_file_paths:
    print(file_path)
```

```python
from ultralytics import YOLO
import cv2
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display as display_image

model = YOLO("yolov8n.pt")

The_bboxes = []
for img_p in sorted_file_paths:
    image_path = img_p
    img = cv2.imread(image_path)
    
    results = model.predict(img)
    
    # Convert OpenCV image to PIL image
    pil_image = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(pil_image)
    font = ImageFont.load_default()
    all_bboxs = []
    
    for r in results:
        boxes = r.boxes
        for box in boxes:
            b = box.xyxy[0]  # get box coordinates in (left, top, right, bottom) format
            c = box.cls
            print(f"bbox {b}, class {c}")
            all_bboxs.append(b)
            draw.rectangle([(b[0], b[1]), (b[2], b[3])], outline="red")
            draw.text((b[0], b[1]), model.names[int(c)], fill="blue", font=font)
    
    data = {
        "bbox":all_bboxs
    }
    The_bboxes.append(data)

    # Display using Pillow
    display_image(pil_image)
```
